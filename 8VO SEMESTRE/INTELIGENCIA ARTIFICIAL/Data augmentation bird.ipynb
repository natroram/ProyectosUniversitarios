```python
from numpy import expand_dims
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
from keras.preprocessing.image import ImageDataGenerator
from matplotlib import pyplot

# cargar la imagen
img = load_img('Bird.jpg')

# convertir a numpy array
data = img_to_array(img)

# expandir 1 dimension en eje 0, as√≠: (sample, alto, largo, canales)
samples = expand_dims(data, 0)
```

```python
img
```

```python
# shift horizontal

# crear una instancia de la clase augmentation generator
datagen = ImageDataGenerator(width_shift_range=[-200,200])

# crear un iterator
it = datagen.flow(samples, batch_size=1)

# generar samples y mostrar
for i in range(9):
    # definir subplot
    pyplot.subplot(330 + 1 + i)
    # generar batch de images
    batch = it.next()
    # convertir a unsigned integers para visualiar
    image = batch[0].astype('uint32')
    # plot raw pixeles
    pyplot.imshow(image)
# mostrar la imagen
pyplot.show()
```

```python
# shift vertical

datagen = ImageDataGenerator(height_shift_range=0.5)
# prepare iterator
it = datagen.flow(samples, batch_size=1)
# generar samples y mostrar
for i in range(9):
    # definir subplot
    pyplot.subplot(330 + 1 + i)
    # generar batch de images
    batch = it.next()
    # convertir a unsigned integers para visualiar
    image = batch[0].astype('uint32')
    # plot raw pixel data
    pyplot.imshow(image)
# mostrar la imagen
pyplot.show()
```

```python
# flip horizontal

datagen = ImageDataGenerator(horizontal_flip=True)
# prepare iterator
it = datagen.flow(samples, batch_size=1)
# generar samples y mostrar
for i in range(9):
    # define subplot
    pyplot.subplot(330 + 1 + i)
    # generar batch de images
    batch = it.next()
    # convertir a unsigned integers para visualiar
    image = batch[0].astype('uint32')
    # plot raw pixel data
    pyplot.imshow(image)
# mostrar la imagen
pyplot.show()
```
